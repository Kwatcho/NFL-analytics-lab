```python
import nfl_data_py as nfl
import pandas as pd

# Get Next Gen Stats for passing
ngs_passing = nfl.import_ngs_data(
    stat_type='passing',
    years=[2023, 2024, 2025]
)

print(f"Data Shape: {ngs_passing.shape}")
print(f"Columns available: {len(ngs_passing.columns)}")

```

```python
# Show all column names
print("All available columns:")
for col in ngs_passing.columns:
    print(f"  - {col}")
```

```python
# Show sample data
print("First few rows:")
print(ngs_passing.head())

```

```python
# Check for release time data
print("Release time data (avg_time_to_throw) is available!")
print(ngs_passing[['player_display_name', 'season', 'week', 'avg_time_to_throw']].head(10))
```

```python
# Filter for meaningful sample sizes
ngs_filtered = ngs_passing[ngs_passing['season'] >= 2025].copy()
```
